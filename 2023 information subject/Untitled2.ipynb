```python
import requests
from bs4 import BeautifulSoup
from tabulate import tabulate

def fetch_typhoon_data(year, typhoon_number):
    try:
        # 입력된 년도와 태풍 번호로 URL 생성
        url = f"https://www.weather.go.kr/w/typhoon/typ-history-pop.do?json=%7B%22typYear%22:{year},%22typSeq%22:{typhoon_number}%7D"

        # 웹 페이지 내용 가져오기
        response = requests.get(url)

        # HTTP 요청이 성공했는지 확인
        if response.status_code == 200:
            # BeautifulSoup를 사용하여 HTML 파싱
            soup = BeautifulSoup(response.text, "html.parser")

            # 데이터를 포함한 테이블 요소 선택
            table = soup.find("table")

            # 중심기압과 최대풍속을 저장할 변수 초기화
            min_pressure = float('inf')  # 중심기압의 최솟값 초기화
            max_wind_speed = 0.0  # 최대풍속의 최댓값 초기화

            # 데이터 행을 선택하고 데이터 값을 추출하여 변수에 저장
            rows = table.find_all("tr")
            for row in rows[2:]:  # 첫 두 행은 열 제목이므로 건너뜁니다.
                columns = row.find_all(["th", "td"])

                # '-' 문자열은 데이터가 없음을 나타내므로 제외합니다.
                if columns[2].get_text(strip=True) != '-':
                    pressure = float(columns[3].get_text(strip=True))
                    min_pressure = min(min_pressure, pressure)
                if columns[4].get_text(strip=True) != '-':
                    wind_speed = float(columns[4].get_text(strip=True))
                    max_wind_speed = max(max_wind_speed, wind_speed)

            # 결과 출력 (년도, 태풍 번호, 중심기압의 최솟값, 최대풍속의 최댓값)
            result = []
            result.append([f"{year}년 {typhoon_number}번 태풍 데이터"])
            result.append([f"중심기압 최솟값(hPa): {min_pressure}"])
            result.append([f"최대풍속 최댓값(m/s): {max_wind_speed}"])

            # 데이터를 표 형식으로 출력
            print(tabulate(result, tablefmt="plain"))
        else:
            print("HTTP 요청이 실패했습니다.")
    except Exception as e:
        # 오류 발생 시 무시하고 아무 작업도 하지 않음
        pass

# 1990부터 2023까지의 년도와 1부터 30까지의 태풍 번호로 데이터를 가져옵니다.
for year in range(1990, 2024):
    for typhoon_number in range(1, 41):
        fetch_typhoon_data(year, typhoon_number)

```
