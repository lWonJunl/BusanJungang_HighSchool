```python
import requests
from bs4 import BeautifulSoup

def fetch_typhoon_data(year, typhoon_number):
    try:
        # 입력된 년도와 태풍 번호로 URL 생성
        url = f"https://www.weather.go.kr/w/typhoon/typ-history-pop.do?json=%7B%22typYear%22:{year},%22typSeq%22:{typhoon_number}%7D"

        # 웹 페이지 내용 가져오기
        response = requests.get(url)

        # HTTP 요청이 성공했는지 확인
        if response.status_code == 200:
            # BeautifulSoup를 사용하여 HTML 파싱
            soup = BeautifulSoup(response.text, "html.parser")

            # 데이터를 포함한 테이블 요소 선택
            table = soup.find("table")

            # 중심기압(hPa)와 최대풍속(m/s)를 저장할 리스트 초기화
            pressure_values = []
            wind_speed_values = []

            # 첫 번째 열이 "일시(KST)"인 경우를 찾아 해당 열의 인덱스를 기억
            headers = table.find_all("th")
            datetime_column_index = None
            for i, header in enumerate(headers):
                if header.get_text(strip=True) == "일시(KST)":
                    datetime_column_index = i

            # 데이터 행을 선택하고 중심기압(hPa)와 최대풍속(m/s) 값을 추출
            rows = table.find_all("tr")
            for row in rows[2:]:  # 첫 두 행은 열 제목이므로 건너뜁니다.
                columns = row.find_all(["th", "td"])

                # '-' 문자열은 데이터가 없음을 나타내므로 제외합니다.
                if columns[2].get_text(strip=True) != '-':
                    pressure_values.append(float(columns[3].get_text(strip=True)))
                if columns[4].get_text(strip=True) != '-':
                    wind_speed_values.append(float(columns[4].get_text(strip=True)))

            # 중심기압(hPa)의 최솟값과 최대풍속(m/s)의 최대값 계산
            min_pressure = min(pressure_values)
            max_wind_speed = max(wind_speed_values)

            # 결과 출력
            print(f"{year}년 {typhoon_number}번 태풍 데이터")
            print(f"중심기압(hPa)의 최솟값: {min_pressure}")
            print(f"최대풍속(m/s)의 최대값: {max_wind_speed}")
        else:
            print("HTTP 요청이 실패했습니다.")
    except Exception as e:
        print(f"오류 발생: {e}")

# 1990부터 2023까지의 년도와 1부터 40까지의 태풍 번호로 데이터를 가져옵니다.
for year in range(1990, 2024):
    for typhoon_number in range(1, 41):
        fetch_typhoon_data(year, typhoon_number)
```
